```python
img1 = cv2.imread('/content/sample_data/images.jpeg')
img_min = img1.min()
img_max = img1.max()
```

```python
normalized_img = (img1 - img_min) / (img_max - img_min) * 255
```

```python
minmax_img = normalized_img.astype('uint8')
cv2_imshow(img1)
cv2_imshow(minmax_img)
```
