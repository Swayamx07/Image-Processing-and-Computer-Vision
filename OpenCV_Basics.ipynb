```python
# Importing the OpenCV library
import cv2
# Reading the image using imread() function
image = cv2.imread('bmw.jpg')
```

```python
# Extracting the height and width of an image
h, w = image.shape[:2]
# Displaying the height and width
print("Height = {}, Width = {}".format(h, w))
```

```python
# Extracting RGB values.
# Here we have randomly chosen a pixel
# by passing in 100, 100 for height and width.
(B, G, R) = image[100, 100]

#Displaying the pixel values
print("R = {}, G= {}, B = {}".format(R, G, B))
```

```python
# Pass the channel to extract (B = 0, G = 1, R = 2)
G = image[100, 100, 1]
print("B = {}".format(G))
```

```python
#You can use this solution if you're using google colab:
#from google.colab.patches import cv2_imshow
#cv2_imshow(img)
from google.colab.patches import cv2_imshow
roi =image[50 : 150, 25 : 270]
cv2_imshow(roi)
cv2.waitKey(0)
```

```python
resize = cv2.resize(image, (500,300))
cv2_imshow(resize)
cv2.waitKey(0)
```

```python
# Calculating the ratio
# w is assumed to be the original width of the image.
# This calculates a scaling factor to make the new width 800 pixels
ratio = 800 / w

# Creating a tuple containing width and height
# h is assumed to be the original height.
# h * ratio calculates the new height, preserving the aspect ratio.
# int() converts the result to an integer, as pixel dimensions must be whole numbers.
# dim becomes a tuple (new_width, new_height) for the cv2.resize function
dim = (800, int(h * ratio))

# Resizing the image
resize_aspect = cv2.resize(image, dim)
cv2_imshow(resize_aspect)
cv2.waitKey(0)
```

```python
# We are copying the original image,
# as it is an in-place operation.
output = image.copy()

# Using the rectangle() function to create a rectangle.
rectangle = cv2.rectangle(output, (200, 100),
                        (100, 50), (255, 0, 0), 2)
cv2_imshow(output)
```

```python
# Copying the original image
output = image.copy()

# Adding the text using putText() function
text = cv2.putText(output, 'OpenCV Demo', (50, 100),
                cv2.FONT_HERSHEY_SIMPLEX, 1, (0, 255, 0), 2)
cv2_imshow(output)
```
